```python
def get_standardized_df(df):
    cols = [df.forecasts.target] + df.forecasts.prediction_columns
    return df[cols]

def evaluate_forecast(df, steps):
    pred_cols = [f'step_pred_{str(i)}' for i in steps]
    df['mse'] = df.apply(lambda row: mean_squared_error([row['ged_sb_dep']] * 36,
                                                        [row[col] for col in pred_cols]), axis=1)
    return df['mse'].mean()
```

```python
df1 = pd.DataFrame.forecasts.read_store(name="orange_pasta")
df1 = get_standardized_df(df1)
df1
```

```python
df2 = pd.DataFrame.forecasts.read_store(name="yellow_pikachu")
df2 = get_standardized_df(df2)
df2
```

```python
def median_ensemble(df1, df2):
    return (df1 + df2) / 2

df_median = median_ensemble(df1, df2)
df_median
```

```python
evaluate_forecast(df1, df1.forecasts.steps),evaluate_forecast(df2, df2.forecasts.steps),evaluate_forecast(df_median, df_median.forecasts.steps)
```
