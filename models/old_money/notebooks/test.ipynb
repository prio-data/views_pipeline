```python
series = test
values = series.all_values()
time_index = series.time_index
components = series.components

# For multivariate series, we can find rows where all values are positive
positive_mask = (values > 0).all(axis=(1, 2))

# Apply the mask to both the values and the time index
filtered_values = values[positive_mask]
filtered_time_index = time_index[positive_mask]
```
