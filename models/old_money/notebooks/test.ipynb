```python
config = {
        "name": "old_money",
        "algorithm": "HurdleRegression",
        "model_clf": "LightGBMModel",
        "model_reg": "LightGBMModel",
        "depvar": "ln_ged_sb_dep",  # IMPORTANT! The current stepshift only takes one target variable! Not compatiable with Simon's code!
        "queryset": "fatalities002_pgm_escwa_drought",
        "level": "pgm",
        "creator": "Xiaolong",
        "sweep": False,
        "steps": [*range(1, 36 + 1, 1)],
        "parameters": {
            "clf": {
                "n_estimators": 200,
            },
            "reg": {
                "n_estimators": 200,
            }
        }
    }

from ingester3.ViewsMonth import ViewsMonth

def get_partitioner_dict(partition, step=36):

    """Returns the partitioner_dict for the given partition."""

    if partition == 'calibration':

        partitioner_dict = {"train":(121,396),"predict":(397,444)} # calib_partitioner_dict - (01/01/1990 - 12/31/2012) : (01/01/2013 - 31/12/2015)

    if partition == 'testing':

        partitioner_dict = {"train":(121,444),"predict":(445,492)} 

    if partition == 'forecasting':

        month_last = ViewsMonth.now().id - 2 # minus 2 because the current month is not yet available. Verified but can be tested by chinging this and running the check_data notebook.

        partitioner_dict = {"train":(121, month_last),"predict":(month_last +1, month_last + 1 + step)}  # is it even meaningful to have a predict partition for forecasting? if not you can remove steps

    # print('partitioner_dict', partitioner_dict) 

    return partitioner_dict
```

```python
depvar = 'ln_ged_sb_dep'
time = df.index.names[0]
level = df.index.names[1]
independent_variables = [c for c in df.columns if c != depvar]
train_start = 121
train_end = 396
test_start = 397
test_end = 444
df_reset = df.reset_index(level=level)
```

```python
target_pos, past_cov_pos = zip(*[(t, p) for t, p in zip(target, past_cov) if (t.values() > 0).any()])
binary_model = LightGBMModel(lags_past_covariates=[-1])
binary_model.fit(binary_target, past_covariates=past_cov)

# Fit positive stage using the regression model
positive_model = LightGBMModel(lags_past_covariates=[-1])
positive_model.fit(target_pos, past_covariates=past_cov_pos)
```

```python
bp = binary_model.predict(n=2, series=target, 
                     past_covariates=[s[independent_variables] for s in series],
                     show_warnings=False)
```

```python
pp = positive_model.predict(n=2, series=target, 
                     past_covariates=[s[independent_variables] for s in series],
                     show_warnings=False)
```

```python
def combine_pred(ts_pred, step):
    preds = []
    for pred in ts_pred:
        df_pred = pred.pd_dataframe()
        df_pred.index = pd.MultiIndex.from_product([df_pred.index, [pred.static_covariates.iloc[0, 0]]])
        df_pred.index.names = ['month_id', 'priogrid_gid']
        df_pred.columns = [f"step_pred_{step}"]
        df_pred = df_pred.loc[slice(test_start, test_end, ), :]
        preds.append(df_pred)

    return pd.concat(preds).sort_index()
```

```python
pred_b = [combine_pred(bp, 1), combine_pred(bp, 2)]
pred_p = [combine_pred(pp, 1), combine_pred(pp, 2)]
```
